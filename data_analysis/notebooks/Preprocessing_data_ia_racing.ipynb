```python
import os
import sys

from os import listdir
from os.path import isfile, join

import pandas as pd    
import numpy as np
import glob

import seaborn as sns
```

```python
path_sigma = "C:/Users/afougere/Git/e1/data_analysis/data/files_data/files_catalog/"

l_circuit = ['waveshare/', 'warehouse/', 'generated/', 'mon_circuit/']

path_catalog_waveshare = path_sigma + l_circuit[0]
path_catalog_warehouse = path_sigma + l_circuit[1]
path_catalog_generated = path_sigma + l_circuit[2]
path_catalog_mon_circuit = path_sigma + l_circuit[3]

fichiers_waveshare = [f for f in listdir(path_catalog_waveshare) if isfile(join(path_catalog_waveshare, f))]
fichiers_warehouse = [f for f in listdir(path_catalog_warehouse) if isfile(join(path_catalog_warehouse, f))]
fichiers_generated = [f for f in listdir(path_catalog_generated) if isfile(join(path_catalog_generated, f))]
fichiers_mon_circuit = [f for f in listdir(path_catalog_mon_circuit) if isfile(join(path_catalog_mon_circuit, f))]

fichiers_waveshare
```

### Transformation des .catalog en .csv

```python
path_sigma_csv = "C:/Users/afougere/Git/e1/data_analysis/data/files_data/catalog_csv/"

path_catalog_csv_wave = path_sigma_csv + l_circuit[0]
path_catalog_csv_ware = path_sigma_csv + l_circuit[1]
path_catalog_csv_gene = path_sigma_csv + l_circuit[2]
path_catalog_csv_mon_circuit = path_sigma_csv + l_circuit[3]

for i in fichiers_waveshare:

    with open(os.path.join(path_catalog_waveshare  + i), 'r', encoding='cp1252') as f1:
        lines = f1.read()
        f2 = open(os.path.join(path_catalog_csv_wave + i + '.csv' ), 'w', encoding='utf-8')
        f2.write(lines)
        #f2.close()
        
for j in fichiers_warehouse:

    with open(os.path.join(path_catalog_warehouse + j), 'r', encoding='cp1252') as f1:
        lines = f1.read()
        f2 = open(os.path.join(path_catalog_csv_ware +  j + '.csv' ), 'w', encoding='utf-8')
        f2.write(lines)
        #f2.close()
        
for f in fichiers_generated:

    with open(os.path.join(path_catalog_generated + f), 'r', encoding='cp1252') as f1:
        lines = f1.read()
        f2 = open(os.path.join(path_catalog_csv_gene + f + '.csv' ), 'w', encoding='utf-8')
        f2.write(lines)
        #f2.close()
        
        
for f in fichiers_mon_circuit:

    with open(os.path.join(path_catalog_mon_circuit + f), 'r', encoding='cp1252') as f1:
        lines = f1.read()
        f2 = open(os.path.join(path_catalog_csv_mon_circuit + f + '.csv' ), 'w', encoding='utf-8')
        f2.write(lines)
        #f2.close()
    
    
#from IPython.core.debugger import Pdb; Pdb().set_trace()
```
